# 1) Loading the radio and X-ray light curves

Radio light curves at three time resolutions, based on the uv-plane analysis:

```python
all_fluxes_perscan = np.genfromtxt('uvplane_fits/flux_target_x1.dat')
all_times_perscan = np.genfromtxt('uvplane_fits/times_x1.dat')
check_fluxes_perscan = np.genfromtxt('uvplane_fits/flux_check_x1.dat')

all_fluxes_x4 = np.genfromtxt('uvplane_fits/flux_target_x4.dat')
all_times_x4 = np.genfromtxt('uvplane_fits/times_x4.dat')
check_fluxes_x4 = np.genfromtxt('uvplane_fits/flux_check_x4.dat')

all_fluxes_x8 = np.genfromtxt('uvplane_fits/flux_target_x8.dat')
all_times_x8 = np.genfromtxt('uvplane_fits/times_x8.dat')
```

```python
# Define the uncertainty in radio flux density per scan from the scaling of the observing time:
dF = 5.0e-3*(len(all_times_perscan)**0.5)
# units are mJy, as that is what we will plot later on
```

Read in the X-ray data from Swift and INTEGRAL:

```python
# Swift
t0_Swift = Time(['2020-03-19T12:12:23.256'], format='isot', scale='utc')

t, dt, r, drp, drm = np.genfromtxt('xray_lightcurves/WT_pipeline.txt', usecols=[0,1,3,4,5], unpack=True)
drm = -1.*drm

t = t*u.second + t0_Swift
```

```python
# INTEGRAL
# Note: we plot the 5-second resolution light curve but the 2-second resolution light curve can be used as well:
# jemx_file = './xray_lightcurves/Rapid_burster_jmx1_3to25keV_2.fits'
jemx_file = './xray_lightcurves/Rapid_burster_jmx1_3to25keV_5s.fits'

hdu1 = fits.open(jemx_file)
data1 = hdu1[2].data
time_jemx_5s_doubles = data1.field('TIME')
rate_jemx_5s_doubles = data1.field('RATE')
error_jemx_5s_doubles = data1.field('ERROR')

hdu1.close()

dt_jemx_5s = 2.5

time_jemx_5s = []
rate_jemx_5s = []
error_jemx_5s = []
for i in range(1, len(time_jemx_5s_doubles)):
    if time_jemx_5s_doubles[i] != time_jemx_5s_doubles[i-1]:
        time_jemx_5s.append(time_jemx_5s_doubles[i])
        rate_jemx_5s.append(rate_jemx_5s_doubles[i])
        error_jemx_5s.append(error_jemx_5s_doubles[i])

time_jemx_5s = np.asarray(time_jemx_5s)
rate_jemx_5s = np.asarray(rate_jemx_5s)
error_jemx_5s = np.asarray(error_jemx_5s)

# Adding the zero-time for the INTEFRAL observation:
t0_INTEGRAL = Time(['2020-03-19T10:23:03.318'], format='isot', scale='utc')
t_INTEGRAL_5s = 24*3600*(time_jemx_5s - time_jemx_5s[0]) * u.second + t0_INTEGRAL
```

## 2) Plotting the light curves together

First, we find the time of bursts from the Swift and INTEGRAL data. These are primarily used for plotting purposes:

```python
# Finding the burst times (start and end) with Swift: count rates exceeding 15 ct/s

burst_starts = []
burst_ends = []
for i in range(1, len(t.mjd)):
    if r[i] > 15. and r[i-1] < 15.:
        burst_starts.append((t.mjd[i]+t.mjd[i-1])/2.)
    if r[i] < 15. and r[i-1] > 15.:
        burst_ends.append((t.mjd[i]+t.mjd[i-1])/2.)

# Finding the burst times (start and end) with INTEGRAL: count rates exceeding 30 ct/s
# Note that we only search in three ranges in the INTEGRAL light curve: those where no Swift data is available.

INT_LIM = 30.

INT_burst_starts = []
INT_burst_ends = []

for i in range(1, 865):
    if rate_jemx_5s[i] > INT_LIM and rate_jemx_5s[i-1] < INT_LIM:
        INT_burst_starts.append((t_INTEGRAL_5s.mjd[i]+t_INTEGRAL_5s.mjd[i-1])/2.)
    if rate_jemx_5s[i] < INT_LIM and rate_jemx_5s[i-1] > INT_LIM:
        INT_burst_ends.append((t_INTEGRAL_5s.mjd[i]+t_INTEGRAL_5s.mjd[i-1])/2.)
        
for i in range(1166, 1568):
    if rate_jemx_5s[i] > INT_LIM and rate_jemx_5s[i-1] < INT_LIM:
        INT_burst_starts.append((t_INTEGRAL_5s.mjd[i]+t_INTEGRAL_5s.mjd[i-1])/2.)
    if rate_jemx_5s[i] < INT_LIM and rate_jemx_5s[i-1] > INT_LIM:
        INT_burst_ends.append((t_INTEGRAL_5s.mjd[i]+t_INTEGRAL_5s.mjd[i-1])/2.)
        
for i in range(1641, len(t_INTEGRAL_5s)):
    if rate_jemx_5s[i] > INT_LIM and rate_jemx_5s[i-1] < INT_LIM:
        INT_burst_starts.append((t_INTEGRAL_5s.mjd[i]+t_INTEGRAL_5s.mjd[i-1])/2.)
    if rate_jemx_5s[i] < INT_LIM and rate_jemx_5s[i-1] > INT_LIM:
        INT_burst_ends.append((t_INTEGRAL_5s.mjd[i]+t_INTEGRAL_5s.mjd[i-1])/2.)   
```

Based on these light curves, we can also calculate the fractional variability:

```python
# Calculting Fvar: based on Vaughan et al. (2003)

def fvar(x, dx):
    # x is the flux density, dx is the flux density 1-sigma uncertainty
    
    xbar2 = np.mean(x)**2
    S2 = sum((x - np.mean(x))**2) / (len(x)-1)
    sigma2 = np.mean(dx**2)
    
    Fvar = ((S2 - sigma2)/xbar2)**0.5

    N = len(x)
    xbar = np.mean(x)

    dFvar = ((((1./(2*N))**0.5) * (sigma2/(xbar2 * Fvar)))**2. + ((sigma2/N) * (1./xbar))**2.)**0.5
    
    return Fvar, dFvar

# Calculating the values for the Rapid Burster with the 1x per scan light curve:
RB_per_scan = fvar(x=1000*np.asarray(all_fluxes_perscan), dx=dF)
```

```python
print('Fractional variability on scan-by-scan basis: F_var = ', np.round(100*RB_per_scan[0],1), '+/-', np.round(100*RB_per_scan[1],1), '%')
```

On shorter time scales, these values are not defined, giving NaN for both values:

```python
print(fvar(x=1000*np.asarray(all_fluxes_x4), dx=dF*2))
print(fvar(x=1000*np.asarray(all_fluxes_x8), dx=dF*(8**0.5)))
print(fvar(x=1000*np.asarray(check_fluxes_perscan), dx=dF))
print(fvar(x=1000*np.asarray(check_fluxes_x4), dx=dF*2))
```

We then also calculate the fluence in the burst times:

```python
NB_fluence = 0. # fluence in non-burst times
B_fluence = 0. # fluence in burst times

N_NB = 0. # number of bins covering the non-burst times
N_B = 0. # number of bins covering the burst times

# Checking the rates and assigning to burst / non-burst times. 
# We use the 1-second resolution of the Swift light curve to convert rate (cts/s) to fluence (cts)
for rate in r:
    if rate > 15:
        B_fluence += rate
        N_B += 1.
    else:
        NB_fluence += rate
        N_NB += 1.

# Calculating the fraction:
print('Fraction of fluence outside bursts:', np.round(NB_fluence/sum(r),3))
print('Fraction of fluence during bursts:', np.round(B_fluence/sum(r),3))
```

We can also calculate the ratio between the mean count rate during burst and non-burst times.

```python
# Mean Swift Non-burst fluence
print('Mean count rate outside bursts:', np.round((NB_fluence/N_NB),2))
print('Mean count rate during bursts:', np.round((B_fluence/N_B),2))
```

## 4) Plotting the difference in flux density as a function of delay:

We finally plot the flux density difference between the burst and non-burst radio measurements, as a function of delay time.

First, we define the fluxes of both burst and non-burst intervals measured from the analysis performed by the '../DATA_ANALYSIS/RADIO/delay_analysis.py' script in the data analysis.

```python
burst_vs_delay = np.genfromtxt('uvplane_fits/burst_vs_delay.dat', unpack=True)
non_burst_vs_delay = np.genfromtxt('uvplane_fits/nonburst_vs_delay.dat', unpack=True)

flux_difference_vs_delay = 1e6*(burst_vs_delay - non_burst_vs_delay)

delays = np.linspace(0, 15*60, 20, endpoint=True)
```

We then scale the uncertainties to the time during and outside bursts:

```python
total_burst_time = sum(((np.asarray(burst_ends) - np.asarray(burst_starts)) * 3600 * 24))
total_time_on_target = 11.*525

flux_error_burst = 5.5 * np.sqrt(total_time_on_target/total_burst_time)
flux_error_non_burst = 5.5 * np.sqrt(total_time_on_target/(total_time_on_target - total_burst_time))

error_flux_difference_vs_delay = (flux_error_burst**2 + flux_error_non_burst**2)**0.5
```

```python
fig = plt.figure(figsize=(cm2inch(3*single_col),cm2inch(1.6*single_col)))
ax = fig.add_subplot(111)

ax.tick_params(labelsize=LS, width=2.5, length=12, axis='both', which='major', pad=5)
ax.tick_params(labelsize=LS, length=8, width=2.5, axis='both', which='minor', pad=5)

ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.get_yaxis().set_tick_params(direction='in', which='both')
ax.get_xaxis().set_tick_params(direction='in', which='both')

ax.errorbar(delays, flux_difference_vs_delay, yerr=error_flux_difference_vs_delay, 
            fmt='sw', ms=10, mew=2, mec='r', elinewidth=2, ecolor='r')
ax.step(delays, 1e6*(burst_vs_delay-non_burst_vs_delay), 'r-', lw=2, where='mid')

ax.plot([-100,1000], [0., 0.], 'k-', lw=2, zorder=-1)
ax.set_xlim(-50, 950)

ax.set_xlabel('Delay [Seconds]', fontsize=FS)
ax.set_ylabel('Flux density difference [uJy]', fontsize=FS)

plt.tight_layout()
plt.savefig('Figure7.pdf')
```

Finally, we can calculate the $\chi^2_\nu$ of this trend versus a constant:

```python
rchi2 = sum(flux_difference_vs_delay) / (error_flux_difference_vs_delay*(len(non_burst_vs_delay) - 1))
print(rchi2)
```
